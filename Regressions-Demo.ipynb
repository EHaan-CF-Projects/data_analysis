```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
from sklearn.model_selection import train_test_split

# render visuzalizations
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [12.0, 8.0]
```

```python
# bunch of random data
def gen_data(size):
    X = np.random.rand(size) * 100
    Y = X ** 2 + 7
    Y = Y + np.random.randn(Y.shape[0]) * 300
    return X, Y
```

```python
X_full, Y_full = gen_data(10)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_full, Y_full, test_size = 0.2, random_state=42)
```

```python
plt.plot(X_train, Y_train, 'bo', markersize=5)
plt.title('Training Set')
```

```python
plt.plot(X_test, Y_test, 'bo', markersize=5)
plt.title('Training Set')
```

```python
coefs_lin = poly.polyfit(X_train, Y_train, 1)
coefs_sq = poly.polyfit(X_train, Y_train, 2)
coefs_ten = poly.polyfit(X_train, Y_train, 3)
```

```python
X_line = np.linspace(0,100)
ffit_ln = poly.polyval(X_line, coefs_lin)
ffit_sq = poly.polyval(X_line, coefs_sq)
ffit_ten = poly.polyval(X_line, coefs_ten)        
```

```python
# clamp values
plt.ylim(min(Y_train - 5000), max(Y_train + 5000))

plt.plot(X_line, ffit_ln, 'g')
plt.plot(X_line, ffit_sq, 'b')
plt.plot(X_line, ffit_ten, 'r')

plt.plot(X_train, Y_train, 'y.', markersize=11)
```
