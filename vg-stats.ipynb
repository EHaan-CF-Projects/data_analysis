### Video Game Sales Data

```python
import numpy as np
import pandas as pd
```

```python
games_frame = pd.read_csv('vgsales.csv')
```

### Assignment Questions:

```python
biggest_publisher = games_frame['Publisher'].mode()
print('The biggest video game publisher is: '+ str(biggest_publisher[0]))
```

```python
biggest_platform = games_frame['Platform'].mode()
print('The most common video game platform is: '+ str(biggest_platform[0]))
```

```python
popularest_genre = games_frame['Genre'].mode()
print('The most popular video game genre is: '+ str(popularest_genre[0]))
```

```python
print('These are the top grossing games:')
games_frame[:20][['Name','Global_Sales']]
```

```python
median_sales = games_frame['NA_Sales'].median()
print('These games had around grossed around ' + str(median_sales) + ', which is the average for North America:')
games_frame.loc[games_frame['NA_Sales'] == 0.08][:10]
```

```python
top_seller = games_frame[:1]
NASales_std = games_frame['NA_Sales'].std()
num_deviations = top_seller['NA_Sales'] / NASales_std

print(str(top_seller['Name'][0]) + 'the top selling game of all time, is ' + str(num_deviations[0]) + ' standard deviations above the average for North American Sales.')
top_seller
```

```python
wii = games_frame['Global_Sales'].loc[games_frame['Platform'] == 'Wii'].mean()
non_wii = games_frame['Global_Sales'].loc[games_frame['Platform'] != 'Wii'].mean()
print( 'The ratio sales from of wii to other game platforms is: ' + str(wii) + ' : ' + str(non_wii))
```

### Three Additional Questions:
- what is the average video game sales for each region: North America, European Union, and Japan?
- what were the most popular games of the 80's?
- how do the genres compare overall?

```python
avg_NA_sales = games_frame['NA_Sales'].mean()
avg_EU_sales = games_frame['EU_Sales'].mean()
avg_JP_sales = games_frame['JP_Sales'].mean()
print('Average North American Sales: ' + str(avg_NA_sales) + '\nAverage European Union Sales: ' + str(avg_EU_sales) + '\nAverage Japan Sales: ' + str(avg_JP_sales))
```

```python
eighties = [1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989]
eighties_games = games_frame[games_frame.Year.isin(eighties)]
print('The top 10 selling games in the eighties were: ')
eighties_games.head(10)
```
